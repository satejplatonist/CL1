```python
import requests
import pandas as pd
import datetime
```

```python
api_key = "39971a16bbe0ea66ad553aca60155735"
```

```python
lat = 18.184135
lon = 74.610764
```

```python
api_url = f"http://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={api_key}"
```

```python
response = requests.get(api_url)
```

```python
weather_data = response.json()
weather_data
```

```python
weather_data.keys()
```

```python
len(weather_data['list'])
```

```python
timestamp = [weather_data['list'][0]['dt'] for items in weather_data['list']]
temperature = [weather_data['list'][0]['main']['temp'] for items in weather_data['list']]
humidity = [weather_data['list'][0]['main']['humidity'] for items in weather_data['list']]
wind_speed = [weather_data['list'][0]['wind']['speed'] for items in weather_data['list']]
weather_description = [weather_data['list'][0]['weather'][0]['description'] for items in weather_data['list']]
```

```python
weather_df = pd.DataFrame({'Timestamp':timestamp, 'Temperature':temperature, 'Humidity':humidity, 'Wind Speed':wind_speed, 'Weather Description':weather_description})
```

```python
weather_df.set_index('Timestamp', inplace=True)
```

```python
weather_df['Temperature'].max()
```

```python
weather_df['Temperature'].min()
```

```python
weather_df.fillna(0,inplace=True)
```

```python
weather_df['Temperature'] = weather_df['Temperature'].apply(lambda x: x - 273.15 if isinstance(x, (int, float)) else x)
```
