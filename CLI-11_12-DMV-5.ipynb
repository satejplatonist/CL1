```python
plt.figure(figsize=(10,8))
plt.pie(data.groupby('shopping_mall').sum()['price'], labels=data.groupby('shopping_mall').sum().index)
plt.title('branch sales price share')
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.pie(data.groupby('category').sum()['price'], labels=data.groupby('category').sum().index)
plt.title('category sales price share')
plt.show()
```

```python
categories = ['Books','Clothing','Cosmetics','Food & Beverage','Toys','Sovenir','Technology','Shoes']
plt.figure(figsize=(10,8))
plt.bar(categories, data.groupby('category').sum()['price'])
plt.title('grouped chart of sales of different categories')
plt.legend()
plt.show()
```
